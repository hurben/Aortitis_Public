```r
library(ggfortify)
library(FactoMineR)
library(factoextra)
library(ggplot2)
```

```r
deg_tpm_file = "../../data/tpm_profile.log2.tsv"
deg_tpm_df = read.csv(deg_tpm_file,sep="\t", header=TRUE, row.names=1)
deg_tpm_df = as.data.frame(t(deg_tpm_df))
```

```r
X = deg_tpm_df[,2:ncol(deg_tpm_df)]
Y = deg_tpm_df[,1]
pca_res <- prcomp(X, center=TRUE)
```

```r
deg_tpm_df$class[deg_tpm_df$class == "1"] <- "case"
deg_tpm_df$class[deg_tpm_df$class == "0"] <- "control"
```

```r
autoplot(pca_res, data = deg_tpm_df, colour = "class")+ ylim(-0.3, 0.5) + xlim(-0.3, 0.4)
```
