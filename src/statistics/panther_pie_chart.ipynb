```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def donout_preprocess(file_dir):
    
    data_df = pd.read_csv(file_dir)
    labels = list(data_df["term"])
    sizes = list(data_df["pie"])
    
    print (labels)
    print (sizes)
    
    return (labels, sizes)

def draw_donut(output_file, labels, sizes):
    
    colors = ['lightcoral', 'yellowgreen',  'lightskyblue', 'gold','plum', 'mediumpurple','dimgray']
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')


    centre_circle = plt.Circle((0,0),0.75,color='black', fc='white',linewidth=3)
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
    
    fig.savefig(output_file)
    
```

```python
DEG_file = '../../data/pie_DEGup_PC.v2.csv'
labels, sizes = donout_preprocess(DEG_file)

output_file = '../../data/pie_DEGup_PC.pdf'
draw_donut(output_file, labels, sizes)
```

```python
DEG_file = '../../data/pie_DEGup_PW.v2.csv'
labels, sizes = donout_preprocess(DEG_file)

output_file = '../../data/pie_DEGup_PW.pdf'
draw_donut(output_file, labels, sizes)

#END UP
```
